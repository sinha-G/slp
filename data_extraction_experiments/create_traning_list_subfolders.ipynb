```python
import os
from collections import defaultdict
import matplotlib.pyplot as plt
import pickle

# New dataset path
dataset_path = 'C:/Users/jaspa/Grant ML/slp/data/segment_2_p_inputs_subfolders'

# Initialize a dictionary to count segments for each character
character_counts = defaultdict(int)

# Traverse the directory structure
for character_folder in os.listdir(dataset_path):
    character_path = os.path.join(dataset_path, character_folder)
    if os.path.isdir(character_path):  # Make sure it's a directory
        for batch_folder in os.listdir(character_path):
            batch_path = os.path.join(character_path, batch_folder)
            if os.path.isdir(batch_path):  # Make sure it's a directory
                for filename in os.listdir(batch_path):
                    if filename.endswith('.npy.gz'):
                        character_counts[character_folder] += 1

# Sort characters by the number of segments
sorted_characters = sorted(character_counts.items(), key=lambda x: x[1], reverse=True)

# Plotting code remains the same...
# [Plotting code here as it remains unchanged]

selected_characters = ['FOX', 'CAPTAIN_FALCON', 'FALCO', 'MARTH', 'SHEIK']
min_segments = 5000  # Adjust as necessary
min_segments = min(character_counts[char] for char in selected_characters)
file_paths = []
labels = {char: idx for idx, char in enumerate(selected_characters)}
label_list = []

# Gather file paths and labels considering the new directory structure
for character in selected_characters:
    count = 0
    character_path = os.path.join(dataset_path, character)
    if os.path.isdir(character_path):
        for batch_folder in os.listdir(character_path):
            batch_path = os.path.join(character_path, batch_folder)
            if os.path.isdir(batch_path):
                for filename in os.listdir(batch_path):
                    if filename.startswith(character) and count < min_segments:
                        full_path = os.path.join(batch_path, filename)
                        file_paths.append(full_path)
                        label_list.append(labels[character])
                        count += 1
                    if count >= min_segments:
                        break
            if count >= min_segments:
                break

print(f"Total files selected: {len(file_paths)}")

save_path = 'C:/Users/jaspa/Grant ML/slp/data/'

# Save file_paths
with open(os.path.join(save_path, 'file_paths.pkl'), 'wb') as f:
    pickle.dump(file_paths, f)

# Save label_list
with open(os.path.join(save_path, 'label_list.pkl'), 'wb') as f:
    pickle.dump(label_list, f)

print("Lists saved successfully.")
```
